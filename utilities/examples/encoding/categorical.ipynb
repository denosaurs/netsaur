```typescript
import { CategoricalEncoder } from "../../src/encoding/categorical.ts";
import { parse } from "jsr:@std/csv@1.0.3/parse";
```

```typescript
// Read the training dataset
const _data = Deno.readTextFileSync("iris.csv");
const data = parse(_data);
data;
```

```typescript
const cat = new CategoricalEncoder();
```

```typescript
const targets = data.map((x) => x[4]);
targets;
```

```typescript
const res = cat.fit(targets).transform(targets, "f32");
```

```typescript
res;
```

```typescript
cat.untransform(res);
```
